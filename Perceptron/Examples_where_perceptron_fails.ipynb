```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
```

```python
and_data = {
    "Input1" : [0,0,1,1],
    "Input2" : [0,1,0,1],
    "Output" : [0,0,0,1]
}
```

```python
or_data = {
    "Input1" : [0,0,1,1],
    "Input2" : [0,1,0,1],
    "Output" : [0,1,1,1]
}
```

```python
xor_data = {
    "Input1" : [0,0,1,1],
    "Input2" : [0,1,0,1],
    "Output" : [0,1,1,0]
}
```

```python
and_data = pd.DataFrame(and_data)
```

```python
or_data = pd.DataFrame(or_data)
```

```python
xor_data = pd.DataFrame(xor_data)
```

```python
a = Perceptron()
b = Perceptron()
c = Perceptron()
```

```python
a.fit(and_data.iloc[:,:2],and_data.iloc[:,-1])
```

```python
b.fit(or_data.iloc[:,:2],or_data.iloc[:,-1])
```

```python
c.fit(xor_data.iloc[:,:2],xor_data.iloc[:,-1])
```

```python
import seaborn as sns
```

```python
sns.scatterplot(
    x=xor_data['Input1'], y=xor_data['Input2'], 
    hue=xor_data['Output'],
    legend='brief', 
)
```

```python
sns.scatterplot(
    x=or_data['Input1'], y=or_data['Input2'], 
    hue=or_data['Output'],
    legend='brief', 
)
```

```python
sns.scatterplot(
    x=and_data['Input1'], y=and_data['Input2'], 
    hue=and_data['Output'],
    legend='brief', 
)
```

```python
w1, w2 = a.coef_[0]
bias = a.intercept_[0]

x1 = np.linspace(-0.5, 1.5, 100)
x2 = (-w1 * x1 - bias) / w2

sns.scatterplot(data=and_data, x='Input1', y='Input2', hue='Output', s=200)
plt.plot(x1, x2, 'k--', label='Decision Boundary')
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.legend()
plt.show()
```

```python
w1, w2 = b.coef_[0]
bias = b.intercept_[0]

x1 = np.linspace(-0.5, 1.5, 100)
x2 = (-w1 * x1 - bias) / w2

sns.scatterplot(data=or_data, x='Input1', y='Input2', hue='Output', s=200)
plt.plot(x1, x2, 'k--', label='Decision Boundary')
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.legend()
plt.show()
```

```python
w1, w2 = c.coef_[0]
bias = c.intercept_[0]

x1 = np.linspace(-0.5, 1.5, 100)
x2 = (-w1 * x1 - bias) / w2

sns.scatterplot(data=xor_data, x='Input1', y='Input2', hue='Output', s=200)
plt.plot(x1, x2, 'k--', label='Decision Boundary (fails)')
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.title('XOR - Not Linearly Separable!')
plt.legend()
plt.show()
```
