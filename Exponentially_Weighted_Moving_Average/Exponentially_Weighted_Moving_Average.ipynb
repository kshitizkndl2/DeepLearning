```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_regression
x, y = make_regression(n_samples=20, n_features=1, n_targets=1,noise = 6)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,)
plt.show()



```

```python
df = pd.DataFrame()
df['x'] = x.reshape(-1) 
df['y'] = y.reshape(-1)

ewma = df['y'].ewm(alpha = 0.9).mean()


df['ewma'] = ewma


```
