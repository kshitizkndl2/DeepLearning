```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_circles

import matplotlib.pyplot as plt

x, y = make_circles(n_samples=1000, noise=0.1, random_state=1)

```

```python
plt.figure(figsize=(10, 10))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.RdYlBu)
plt.show()



```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)



```

```python
from tensorflow.keras import callbacks
callback = callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model1  = Sequential()

model1.add(Dense(128, input_shape=(2,), activation='relu'))
model1.add(Dropout(0.3))
model1.add(Dense(64, activation='relu'))
model1.add(Dropout(0.4))
model1.add(Dense(1, activation='linear'))

model1.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
```

```python
model1.fit(x_train, y_train, epochs=2000, validation_data=(x_test, y_test),callbacks  = callback)
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x_train, y_train, clf=model1, legend=2)


```

we can see that using the activation function can help to improve the neural network performance.
